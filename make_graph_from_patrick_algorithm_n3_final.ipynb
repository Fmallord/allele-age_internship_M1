## 3 - Algorithm inferring "true" allele age

```python
Ne=7300
myfile1 = open("/media/francoismallord/USB DISK/chr1_for_real_age/n3/n3_estimate_1000_allele_comb1.txt", "w")
myfile2 = open("/media/francoismallord/USB DISK/chr1_for_real_age/n3/n3_real_1000_allele_comb1.txt", "w")
real=0             # initialisation of "true" allele age
estimate=0         # initialisation of estimate allele age
age_c=[]
age_d=[]
t_c=0                # initialisation of t_c
t_d=0               # initialisation of t_d
Estimate=[]          #list that will store the inferred allele ages
Real=[]              #list that will store the "true" allele ages
T_c=[]
T_d=[]
est_age3f=[i for i in est_age3 if i[1]=='C']            #we only consider the inferred allele age from cle files giving a clock model (here, C=combined clock)
for i in est_age3f:
    estimate=2*Ne*float(i[7])              #pop-scaled inferred allele age (mode of CP)
    Estimate+=[estimate]
    myfile1.write("%s\n" % estimate)
    marker_spe=[a for a in est_age2f if a[0]==i[0]] #we focus only on all TMRCA from the haplotype pairs for a given marker i[0]
    pairs_c=[j for j in marker_spe if j[7]=='1'] #all conc pairs for marker i[0]
    pairs_d=[k for k in marker_spe if k[7]=='0'] #all disc pairs for marker i[0]
    for i in pairs_c:
        age_c+=[float(i[14])]
    t_c=max(age_c)                     #the oldest TMRCA in concordant pairs picked to estimate allele age
    T_c+=[t_c]
    for i in pairs_d:
        age_d+=[float(i[14])]
    t_d=min(age_d)                     #the youngest TMRCA in concordant pairs picked to estimate allele age
    T_d+=[t_d]
    real=math.sqrt(t_c*t_d)            # "true" age = geomemtric mean of t_c and t_d
    Real+=[real]
    myfile2.write("%s\n" % real)
    age_c=[]
    age_d=[]
myfile1.close()
myfile2.close()
```
